### Apply()

```python
def categorize_age(age):
    if pd.isnull(age):  # Kiểm tra nếu giá trị là NaN (null)
        return np.nan  # Trả về NaN nếu giá trị là null
    elif age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

# Áp dụng hàm categorize_age lên cột 'Age'
df['Age_Category'] = df['Age'].apply(categorize_age)

# Hiển thị kết quả
print(df[['Name', 'Age', 'Age_Category']].head())
```
